```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
from dotenv import load_dotenv
project_root = os.path.abspath("..")
sys.path.insert(0, project_root)
```

```python
load_dotenv()
BASE_PATH = os.getenv("BASE_PATH")
df = pd.read_parquet(f"{BASE_PATH}/data/processed/clean_delivery_time.parquet")
df.head()
```

```python
traffic_map = {'Low': 0, 'Medium': 1, 'High': 2}
df['Traffic_Level'] = df['Traffic_Level'].map(traffic_map)
```

```python
df['Traffic_Level'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Weather', 'Time_of_Day', 'Vehicle_Type'], drop_first=True)
```

```python
df.head()
```

```python
df.to_parquet(f"{BASE_PATH}/data/processed/feature_engineering_clean_delivery_time.parquet")
```
